```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
LOVE= pd.read_csv("C:/Users/Lenovo/Downloads/love1.csv")
```

```python
LOVE.head()
```

```python
LOVE= LOVE.drop(['Country Code','Indicator Name','Indicator Code'], axis=1)
```

```python
chosen_country = 'Afghanistan'


country_data = LOVE[LOVE['Country Name'] == chosen_country]


years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

plt.figure(figsize=(5, 3))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')


plt.show()

     
```

```python

plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.show()
```
